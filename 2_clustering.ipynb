````python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
````

````python
df = pd.read_csv('./data/retailmax.csv')
````

````python
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
````

````python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
````

````python
type(scaled_features)
````

````python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
````

````python
# Función para calcular la Suma de Cuadrados Dentro del Cluster (WCSS)
def calcular_wcss(datos):
    wcss = []
    for n in range(1, 11):
        kmeans = KMeans(n_clusters=n, init='k-means++', max_iter=300, n_init=10, random_state=42)
        kmeans.fit(datos)
        wcss.append(kmeans.inertia_)
    return wcss

# Calcular el WCSS para diferentes números de clusters
wcss = calcular_wcss(scaled_features)

# Graficar el método del codo
plt.plot(range(1, 11), wcss)
plt.title('Método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
````

## Resumen Ejecutivo

A través del análisis de clustering aplicando el algoritmo K-Means, hemos identificado **5 segmentos distintos de clientes** en nuestra base de datos. Cada segmento presenta características demográficas y de comportamiento de compra únicas que requieren estrategias de marketing diferenciadas. Este reporte presenta recomendaciones específicas para maximizar el engagement y las conversiones en cada grupo.

---

## Segmentos Identificados y Estrategias Recomendadas

### 🎯 **Cluster 0: "Premium Enthusiasts" (Alto Valor)**
**Perfil:**
- Edad promedio: 33 años
- Ingreso anual: $86,540
- Puntuación de gasto: 82.13/100
- Composición: 54% mujeres

**Características clave:** Este es nuestro segmento más valioso. Son profesionales jóvenes con alto poder adquisitivo que disfrutan gastar en nuestros productos.

**Recomendaciones:**
1. **Programa VIP exclusivo:** Crear una membresía premium con beneficios especiales, acceso anticipado a nuevos productos y eventos exclusivos.
2. **Marketing de experiencias:** Campañas enfocadas en experiencias premium y productos de alta gama.
3. **Personalización avanzada:** Implementar recomendaciones de productos basadas en IA y comunicación uno a uno.
4. **Retención prioritaria:** Este grupo debe ser nuestro foco principal de retención. Invertir en programas de lealtad con recompensas significativas.

---

### 💼 **Cluster 1: "Conservadores Adinerados" (Potencial Sin Explotar)**
**Perfil:**
- Edad promedio: 36 años
- Ingreso anual: $89,520
- Puntuación de gasto: 18.00/100
- Composición: 55% mujeres

**Características clave:** Tienen el ingreso más alto de todos los segmentos, pero gastan muy poco. Representan una oportunidad enorme.

**Recomendaciones:**
1. **Estrategia de confianza:** Crear campañas que destaquen calidad, durabilidad y valor a largo plazo de los productos.
2. **Educación del producto:** Contenido informativo que demuestre el ROI y beneficios de invertir en productos premium.
3. **Pruebas sin riesgo:** Ofrecer garantías extendidas, devoluciones fáciles y períodos de prueba para reducir el riesgo percibido.
4. **Marketing de referencias:** Estos clientes responden bien a testimonios y recomendaciones de pares profesionales.
5. **Ofertas estratégicas:** Promociones exclusivas en fechas clave (fin de año fiscal, bonos profesionales).

---

### 👩 **Cluster 2: "Mujeres Pragmáticas" (Segmento Estable)**
**Perfil:**
- Edad promedio: 50 años
- Ingreso anual: $49,230
- Puntuación de gasto: 40.07/100
- Composición: 100% mujeres

**Características clave:** Segmento exclusivamente femenino de mediana edad con ingresos moderados y comportamiento de compra equilibrado.

**Recomendaciones:**
1. **Marketing de valor:** Enfatizar relación calidad-precio y productos multifuncionales.
2. **Contenido relevante:** Campañas que resuenen con sus intereses y etapa de vida (familia, hogar, bienestar).
3. **Programas de lealtad:** Puntos por compra y descuentos progresivos que premien la consistencia.
4. **Canales tradicionales:** Este grupo responde mejor a email marketing y catálogos que a redes sociales.
5. **Bundles y paquetes:** Ofertas que maximicen el valor percibido.

---

### 🌟 **Cluster 3: "Jóvenes Aspiracionales" (Alto Potencial a Futuro)**
**Perfil:**
- Edad promedio: 25 años
- Ingreso anual: $39,720
- Puntuación de gasto: 61.20/100
- Composición: 59% mujeres

**Características clave:** Los clientes más jóvenes con ingresos modestos pero alta propensión al gasto. Gastan un porcentaje significativo de sus ingresos en nuestros productos.

**Recomendaciones:**
1. **Marketing en redes sociales:** Campañas agresivas en Instagram, TikTok y plataformas donde este grupo pasa tiempo.
2. **Influencer marketing:** Colaboraciones con micro-influencers que resuenen con este grupo demográfico.
3. **Opciones de pago flexible:** Implementar "compra ahora, paga después" y planes de financiamiento.
4. **Productos de entrada:** Líneas de productos accesibles que permitan construir lealtad temprana.
5. **Comunidad y pertenencia:** Crear una comunidad de marca donde puedan interactuar y sentirse parte de algo más grande.
6. **Inversión a largo plazo:** Aunque sus ingresos son moderados ahora, son profesionales jóvenes cuyo poder adquisitivo crecerá.

---

### 👨‍🦳 **Cluster 4: "Maduros Tradicionales" (Segmento Conservador)**
**Perfil:**
- Edad promedio: 56 años
- Ingreso anual: $53,690
- Puntuación de gasto: 36.77/100
- Composición: 100% hombres

**Características clave:** Hombres mayores con ingresos moderados y comportamiento de compra conservador.

**Recomendaciones:**
1. **Marketing directo:** Email marketing personalizado y correo físico funcionan mejor que anuncios digitales.
2. **Enfoque en funcionalidad:** Destacar utilidad, durabilidad y practicidad sobre tendencias.
3. **Servicio al cliente excepcional:** Este grupo valora enormemente la atención personal y el servicio post-venta.
4. **Promociones estacionales:** Ofertas en fechas específicas (Día del Padre, fin de año).
5. **Comunicación clara:** Mensajes directos sin jerga, con información completa del producto.

---

## 🎯 Prioridades Estratégicas

### Prioridad Alta
1. **Retención de Cluster 0 (Premium Enthusiasts):** Representan el 26% de clientes pero generan el mayor valor por cliente.
2. **Conversión de Cluster 1 (Conservadores Adinerados):** Tienen el mayor potencial de incremento en volumen de compra.
3. **Desarrollo de Cluster 3 (Jóvenes Aspiracionales):** Inversión en lealtad temprana para crecimiento a largo plazo.

### Prioridad Media
4. **Estabilización de Cluster 2 (Mujeres Pragmáticas):** Mantener su comportamiento de compra consistente.
5. **Optimización de Cluster 4 (Maduros Tradicionales):** Incrementar levemente frecuencia de compra.
